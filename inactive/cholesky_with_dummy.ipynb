```python
#cholesky factorization of A (xtx)
c_and_lower = scipy.linalg.cho_factor(xtx, lower = True)
c_and_lower
```

```python
#solving the linear equations of Ax = b, b is xty and A is the cholesky factorization we got above
beta_hat_ch = scipy.linalg.cho_solve(c_and_lower, x.transpose() @ y)
beta_hat_ch
```

```python
#calculating cholesky value
L = np.linalg.cholesky(xtx)
L
```

```python
#using the property L^t beta = l
sm_l = L.transpose() @ beta_hat_ch
sm_l
```

## Variance and Standard Error of $\hat\beta$

```python
#variance
var_ch = (y.transpose() @ y - sm_l.transpose() @ sm_l)/(x.shape[0] - x.shape[1])
var_ch
```

```python
#standard error
se_ch = np.sqrt(var_ch * np.diag(scipy.linalg.inv(L @ L.transpose())))
se_ch
```
